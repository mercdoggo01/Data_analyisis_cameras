## Step 2: Data preparation & cleaning

```python
All_cameras_df = pd.read_csv('camera_dataset.csv')
```

### Now since dataset is very large we use Ipython to view the whole thing

```python

from IPython.display import display
with pd.option_context('display.max_rows', 2000):
    display(All_cameras_df)
```

### Here we get 1038 rows and 13 columns

### Here we can se that the rows are not named appropriately for usage (i.e it contains spaces)

```python
# We fix it then:
All_cameras_df['Release_date'] = All_cameras_df['Release date']
All_cameras_df['Max_resolution'] = All_cameras_df['Max resolution']
All_cameras_df['Low_resolution'] = All_cameras_df['Low resolution']
All_cameras_df['Effective_pixels'] = All_cameras_df['Effective pixels']
All_cameras_df['Zoom_W'] = All_cameras_df['Zoom wide (W)']
All_cameras_df['Zoom_T'] = All_cameras_df['Zoom tele (T)']
All_cameras_df['Normal_focus_range'] = All_cameras_df['Normal focus range']
All_cameras_df['Macro_focus_range'] = All_cameras_df['Macro focus range']
All_cameras_df['Storage_included'] = All_cameras_df['Storage included']
All_cameras_df['Weight_inc_batteries'] = All_cameras_df['Weight (inc. batteries)']
New_All_cameras_df = All_cameras_df.drop(['Release date','Max resolution','Low resolution','Effective pixels','Zoom wide (W)','Zoom tele (T)','Normal focus range','Macro focus range', 'Storage included','Weight (inc. batteries)'], axis=1 )
```

```python
New_All_cameras_df
```

```python
# Now we check for missing or inconsistent data
New_All_cameras_df.isna()
```

### Since we find that using isna() we are getting false values hence there is no missing and inconsistent data.
### 1. Now at many places 0.0 is the given value so we assume them as low quality cameras and treat it as a min value
### 2. Secondly now we just need to arrange the values in ascending order in terms of Release_date

```python
New_All_cameras_df = New_All_cameras_df.sort_values('Release_date', ascending=True)
New_All_cameras_df
```

# Step 3: Exploratory analysis & visualization

## General statistics of the dataset (i.e mean , count, max , min) for all columns

```python
New_All_cameras_df.describe()
```

### Now we will check the quality increase of the cameras over the years


```python
# we will use matplotlib here extensively with focus on pyplot 
plt.plot(New_All_cameras_df.Release_date, New_All_cameras_df.Max_resolution)
plt.plot(New_All_cameras_df.Release_date, New_All_cameras_df.Low_resolution)


plt.xlabel('Year')
plt.ylabel('Maximum resolution')

plt.title("Quality increase over the years")
plt.legend(['Max_resolution', 'Min_resolution']);
```

### As you can se that the first breakthrough in maximum resolution occuered in 1994-1996 and there after has been constantly improved upon as for minimum resolution it occured in 1996-1998


### Now we will analyze the weight of cameras over the years

```python
plt.figure(figsize=(20,5))
sns.set_style("darkgrid")
plt.plot(New_All_cameras_df.Release_date,New_All_cameras_df.Weight_inc_batteries,'s-b')
plt.xlabel('Year')
plt.ylabel('Weight (in gms)')

plt.title("Weight over the years")
plt.legend(['Weights']);
```

### As we can observe here that there is a general trend of decrease in the weights of the cameras overtime though there have been instances of cameras with heigher weights releasing annualy

```python
plt.figure(figsize=(12, 6))
plt.title('Scatter plot of dimensions & weights')

sns.scatterplot(x=New_All_cameras_df.Dimensions, 
                y=New_All_cameras_df.Weight_inc_batteries, 
                s=100);
```

### In this scatterplot we can se that the most favourable choice of camera is the ones with weight around 750gms and with dimension of <150cms

```python
plt.figure(figsize=(15,8))
sns.barplot(x= New_All_cameras_df.Release_date, y=New_All_cameras_df.Price, data=New_All_cameras_df);
```

### Price increase is not much since 2001 where there was a price spike. Before 2001 same pattern was observed wherein the price peaked at 1996 there after it decreased and was more or less constant till 2001

### Q 4: How many camera models were manufactured yearwise show in any plot

```python
years = All_cameras_df.Release_date.unique()
#sorting done in order to arrange yearwise from 1994 to 2007
years.sort()
years
```

```python
manufacturing=[]
for i in years:
    manufacturing.append(New_All_cameras_df[New_All_cameras_df.Release_date == i].Model.count())
```

```python
manufacturing
```

```python
plt.figure(figsize=(15, 6))
plt.plot(years, manufacturing)
plt.xlabel('Years')
plt.ylabel('production');
```
